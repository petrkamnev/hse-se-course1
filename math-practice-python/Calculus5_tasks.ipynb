```python
import numpy as np
import sympy
from sympy import nsolve
import matplotlib.pyplot as plt
%matplotlib inline
```

### –ó–∞–¥–∞—á–∞ 1
–†–µ—à–∏—Ç—å —É—Ä–∞–≤–Ω–µ–Ω–∏–µ $ùë•^4‚àíe^x=0$ –ø—Ä–∏ $-2 \le x \le 2$, –ø–æ—Å—Ç—Ä–æ–∏—Ç—å –≥—Ä–∞—Ñ–∏–∫ –ª–µ–≤–æ–π —á–∞—Å—Ç–∏, –æ—Ç–º–µ—Ç–∏—Ç—å —Ç–æ—á–∫–∏ –ø–µ—Ä–µ—Å–µ—á–µ–Ω–∏—è —Å –æ—Å—å—é –û–•.

```python
x = sympy.Symbol('x')
f = x**4 - sympy.exp(x)
X = np.linspace(-2, 2, 100)
plt.plot(X, X**4 - np.exp(X))
plt.ylim(-1, 1)
ax = plt.gca() 
ax.spines['right'].set_color('none') 
ax.spines['top'].set_color('none') 
ax.spines['left'].set_position(('data', 0)) 
ax.spines['bottom'].set_position(('data', 0))
for a, b in [(-1, 'A'), (1.5, 'B')]:
    ans = nsolve(f, x, a)
    display(ans)
    plt.scatter(ans, 0)
    ax.annotate(b, xy=(ans, 0), xycoords='data', xytext=(ans - 0.15, 0.03))
```

### –ó–∞–¥–∞—á–∞ 2.
–†–µ—à–∏—Ç—å —É—Ä–∞–≤–Ω–µ–Ω–∏–µ $\sin^2(2x)=\ln x$ –ø—Ä–∏ $0< x \le 3$. –ü–æ—Å—Ç—Ä–æ–∏—Ç—å –≥—Ä–∞—Ñ–∏–∫–∏ –ª–µ–≤–æ–π –∏ –ø—Ä–∞–≤–æ–π —á–∞—Å—Ç–∏, –æ—Ç–º–µ—Ç–∏—Ç—å —Ç–æ—á–∫—É –ø–µ—Ä–µ—Å–µ—á–µ–Ω–∏—è –≥—Ä–∞—Ñ–∏–∫–æ–≤.

```python
x = sympy.Symbol('x')
f1 = (sympy.sin(2*x))**2
f2 = sympy.ln(x)
X = np.linspace(0.001, 3, 100)
plt.plot(X, [f1.subs(x, i).evalf() for i in X])
plt.plot(X, [f2.subs(x, i).evalf() for i in X])
plt.ylim(-1, 1)
ax = plt.gca() 
ax.spines['right'].set_color('none') 
ax.spines['top'].set_color('none') 
ax.spines['left'].set_position(('data', 0)) 
ax.spines['bottom'].set_position(('data', 0))
for a, b in [(1.25, 'A'), (2, 'B'), (2.5, 'C')]:
    ans = nsolve(f2 - f1, x, a)
    display(ans)
    plt.scatter(ans, f1.subs(x, ans).evalf())
    ax.annotate(b, xy=(ans, f1.subs(x, ans).evalf()), xycoords='data', xytext=(ans + 0.05, f1.subs(x, ans).evalf() + 0.1))
```

### –ó–∞–¥–∞—á–∞ 3.
–†–µ—à–∏—Ç—å —É—Ä–∞–≤–Ω–µ–Ω–∏–µ $\sin^2(2x^2-7x+6)=3/4$ –Ω–∞ $[3, 4]$. –ü–æ—Å—Ç—Ä–æ–∏—Ç—å –≥—Ä–∞—Ñ–∏–∫–∏ –ª–µ–≤–æ–π –∏ –ø—Ä–∞–≤–æ–π —á–∞—Å—Ç–∏, –æ—Ç–º–µ—Ç–∏—Ç—å —Ç–æ—á–∫–∏ –ø–µ—Ä–µ—Å–µ—á–µ–Ω–∏—è –≥—Ä–∞—Ñ–∏–∫–æ–≤.

```python
x = sympy.Symbol('x')
f = (sympy.sin(2*x*x - 7*x + 6))**2
X = np.linspace(3, 4, 100)
plt.plot(X, [f.subs(x, i).evalf() for i in X])
plt.plot(X, [3/4 for i in range(len(X))])
plt.ylim(0, 1)
ax = plt.gca() 
ax.spines['right'].set_color('none') 
ax.spines['top'].set_color('none') 
ax.spines['left'].set_position(('data', 3)) 
ax.spines['bottom'].set_position(('data', 0))
for a, b in [(3, 'A'), (3.4, 'B'), (3.5, 'C'), (3.7, 'D')]:
    ans = nsolve(f - 3/4, x, a)
    display(ans)
    plt.scatter(ans, f.subs(x, ans).evalf())
    ax.annotate(b, xy=(ans, f.subs(x, ans).evalf()), xycoords='data', xytext=(ans + 0.05, f.subs(x, ans).evalf() + 0.1))
```

### –ó–∞–¥–∞—á–∞ 4.
–†–µ—à–∏—Ç—å —É—Ä–∞–≤–Ω–µ–Ω–∏–µ ${\rm tg}^4(2x^2+1)-10=0$ –Ω–∞ –∏–Ω—Ç–µ—Ä–≤–∞–ª–µ $[0, 0.2]$ –º–µ—Ç–æ–¥–æ–º –¥–∏—Ö–æ—Ç–æ–º–∏–∏. –ì—Ä–∞—Ñ–∏–∫ —Å—Ç—Ä–æ–∏—Ç—å –Ω–µ –Ω—É–∂–Ω–æ.

```python
x = sympy.Symbol('x')
f = (sympy.tan(2*x**2 + 1))**4 - 10
nsolve(f, x, [0, 0.2], solver='bisect', verify=False)
```

### –ó–∞–¥–∞—á–∞ 5.
–†–µ—à–∏—Ç—å —Å–∏—Å—Ç–µ–º—É —É—Ä–∞–≤–Ω–µ–Ω–∏–π 
$$
\left\{\begin{matrix}
(x - 1)^2 + (y - 2)^2 + (z + 3)^2 = 25\\
x + y + z = 0\\
-2x + 3y - z = 7
\end{matrix}\right.
$$
–Ω–∞—á–∞–ª—å–Ω–æ–µ –ø—Ä–∏–±–ª–∏–∂–µ–Ω–∏–µ (0, 0, 0).
–ü—Ä–æ–≤–µ—Ä–∏—Ç—å –ø–æ–¥—Å—Ç–∞–Ω–æ–≤–∫–æ–π.

```python
x, y, z = sympy.symbols('x y z')
system = [(x - 1)**2 + (y - 2)**2 + (z + 3)**2 - 25, x + y + z, -2*x + 3*y - z - 7]
ans = nsolve(system, (x, y, z), (0, 0, 0))
display(ans)
for i in system:
    display(i.subs({x: ans[0], y: ans[1], z : ans[2]}))
```
